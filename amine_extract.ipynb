```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
df = pd.read_csv('/Users/emmaking-smith/Moonshot/noisy_amides.csv', index_col=0)
df = df.reset_index(drop=True)
```

```python
print(df.loc[0, 'SMILES'])
peptides = [Chem.MolFromSmiles(s) for s in df['SMILES']]
peptides[0]
```

```python
acid = 'Clc1cc2c([C@H](C(Nc(cnc3)c4c3cccc4)=O)CN(CC=O)C2)cc1'
acid = Chem.MolFromSmarts(acid)
```

```python
amines = []
for mol in peptides:
    amine = AllChem.DeleteSubstructs(mol, acid)
    amine.UpdatePropertyCache()
    Chem.GetSymmSSSR(amine)
    amines.append(amine)
```
